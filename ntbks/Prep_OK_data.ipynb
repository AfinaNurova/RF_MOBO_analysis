# Cleaning the baseline APARC data

```python
import scipy.io
roi_mat = scipy.io.loadmat('Datos_ADNI_Afina/roi_aparc_a2009s.mat', struct_as_record=False, squeeze_me=True)
```

```python
roi_mat.keys()
```

```python
tab_data = roi_mat["tab"]  # Extract the variable

print(type(tab_data))  # Check its type
print(tab_data.__dict__.keys())  # List all field names
```

```python
roi_ap = tab_data.__dict__
roi_ap
```

```python
import re

session_number = [ re.search('\d+', x)[0]  for x in roi_ap['session'] ]
roi_ap['session_number'] = session_number
```

```python
len(session_number)
```

```python
len(np.unique(roi_ap['subj']))
```

```python
roithickness = roi_ap['roithickness']
print(roithickness.shape)
```

```python
roinames = list(roi_ap['roinames'])
len(roinames)
```

```python
subject = list(roi_ap['subj'])
len(subject)
```

```python
joint_df = pd.DataFrame({'subject': subject, 'session_number': session_number, **dict(zip(roinames, roithickness.T ))})
print(joint_df.shape)
joint_df.head()
```

```python
baseline_joint_df = joint_df[joint_df['session_number'] == '00']
print(baseline_joint_df.shape)
baseline_joint_df.head()
```

```python
baseline_joint_df = baseline_joint_df.assign(RID=baseline_joint_df['subject'].apply(lambda x: re.search(r'S0*([1-9]\d*)', x).group(1) if pd.notna(x) else None))
baseline_joint_df.head()
```

```python
baseline_joint_df.to_pickle('baseline_roi_aparc_df.pkl')
```

```python
baseline_joint_df.to_parquet('baseline_roi_aparc_df.parquet', engine='pyarrow')
```
