```python
# === 1. Folder containing LD logs (SingleGP) ===
folder_path = "./Afina_results_very_new/logs/SingleGP"

# Load all Parquet files starting with 'LD_'
file_paths = glob.glob(os.path.join(folder_path, "LD_*.parquet"))
file_paths = [f for f in file_paths if not os.path.basename(f).startswith("LD_NEURO")]

print(file_paths)

# === 2. Load and concatenate all files ===
all_logs = [pd.read_parquet(fp) for fp in file_paths]
combined_df = pd.concat(all_logs, ignore_index=True)

# Save combined data
combined_df.to_parquet("./LD_SGP_combined_logs.parquet")
# combined_df.to_csv("./LD_combined_logs.csv", index=False)

```
